```python
import os
from os import walk
import numpy as np
import matplotlib as mpl
mpl.use('Agg')
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy.linalg as linalg
from theano import shared

from PIL import Image

import theano
import theano.tensor as T
```

```python
filepath = 'Fei_256/'
names = sorted(os.listdir(filepath))
f=mpl.pyplot.imread(filepath+names[0]).reshape((1,-1))
Ims=np.array(f,dtype=np.uint8)
for i in range(1,len(names)):
    f=mpl.pyplot.imread(filepath+names[i]).reshape((1,-1)) 
    Ims=np.append(Ims,f,axis=0)
```

```python
Ims = Ims.astype(np.float32)
X_mn = np.mean(Ims, 0)
X = Ims - np.repeat(X_mn.reshape(1, -1), Ims.shape[0], 0)
```

```python
d1=T.fvector('arg1')
d2=T.fvector('arg2')
XX=T.fmatrix('arg3')
d3=T.fvector('arg4')
mat_vec1=T.dot(XX,d1)
mat_vec2=T.dot(XX.T,mat_vec1)
mat_vec3=T.dot(XX,d2)
vec_vec1=T.dot(d2.T,d1)

cost=-T.dot(mat_vec1.T,mat_vec1)# used in the 1st component
lamda=T.dot(mat_vec3.T,mat_vec3)
vec_vec2=lamda*d2*vec_vec1

grad=theano.gradient.grad(cost,d1)
grad_p1=-2*mat_vec2+2*vec_vec2
grad_p12=d3+2*vec_vec2

calc_grad=theano.function([d1,XX],grad)
calc_cost=theano.function([d1,XX],cost)
```

```python
epsilon=1.e-05
learn_rate=0.004
count=0
stop_iteration=1500
#D=np.zeros((65536,16),dtype=np.float32)
```

```python
old_d=np.random.randn(65536).astype(np.float32)
old_d=old_d/linalg.norm(old_d)
dervtv=calc_grad(old_d,X)
print(calc_cost(old_d,X))
dervtv=dervtv/linalg.norm(dervtv)
```

```python
while count<stop_iteration:
    new_d=old_d-dervtv*learn_rate
    new_d=new_d/linalg.norm(new_d)
    if linalg.norm(new_d-old_d)<epsilon:
        break
    dervtv=calc_grad(new_d,X)
#    if linalg.norm(dervtv)<epsilon:
#        break
    dervtv=dervtv/linalg.norm(dervtv)
    old_d=new_d
    count=count+1
    last_cost=calc_cost(new_d,X)
D[:,0]=new_d
print(count)
```

```python
temp_drv=shared(np.zeros((65536,),dtype=np.float32))
calc_p1=theano.function([d1,d2,XX],grad_p1,updates=[(temp_drv,grad_p1)])
calc_p2=theano.function([d3,d2,d1,XX],grad_p12,updates=[(temp_drv,temp_drv+2*vec_vec2)])
```

```python
old_d=np.random.randn(65536).astype(np.float32)
old_d=old_d/linalg.norm(old_d)
count=0
for j in range(0,3):
    known_d=D[:,j]
    if j==0:
        calc_p1(old_d,known_d,X)
    else:
        calc_p2(temp_drv.get_value(),known_d,old_d,X)
    print(temp_drv.get_value())
```

```python
dervtv=temp_drv.get_value()
dervtv=dervtv/linalg.norm(dervtv)
```

```python
while count<stop_iteration:
    new_d=old_d-dervtv*learn_rate
    new_d=new_d/linalg.norm(new_d)
    if linalg.norm(new_d-old_d)<epsilon:
        break
    for j in range(0,3):
        known_d=D[:,j]
        if j==0:
            calc_p1(new_d,known_d,X)
        else:
            calc_p2(temp_drv.get_value(),known_d,new_d,X)
    dervtv=temp_drv.get_value()
    dervtv=dervtv/linalg.norm(dervtv)
    old_d=new_d
    count=count+1
    print(count)
```

```python
for i in range(5,7):
    old_d=np.random.randn(65536).astype(np.float32)
    old_d=old_d/linalg.norm(old_d)
    count=0
    known_d=D[:,0]
    calc_p1(old_d,known_d,X)
    for j in range(1,i):
        known_d=D[:,j]
        calc_p2(temp_drv.get_value(),known_d,old_d,X)

    dervtv=temp_drv.get_value()
    dervtv=dervtv/linalg.norm(dervtv)

    while count<stop_iteration:
        new_d=old_d-dervtv*learn_rate
        new_d=new_d/linalg.norm(new_d)
        if linalg.norm(new_d-old_d)<epsilon:
            break
        known_d=D[:,0]
        calc_p1(new_d,known_d,X)
        for j in range(1,i):
            known_d=D[:,j]
            calc_p2(temp_drv.get_value(),known_d,new_d,X)
        dervtv=temp_drv.get_value()
        dervtv=dervtv/linalg.norm(dervtv)
        old_d=new_d
        count=count+1
        print(i,count)
#last_cost=calc_cost(new_d,X)
    D[:,i]=new_d
```
